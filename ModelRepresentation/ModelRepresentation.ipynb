#### Implementing the $f_{w,b}$ for linear regression with one variable.

##### Housing price prediction.



```python
import numpy as np
import matplotlib.pyplot as plt
```

Prediction using only two data points.
a house with 1000 square feet(sqft) sold for $300,000 and a house with 2000 square feet sold for $500,000. These two points will constitute our data or training set. The units of size are 1000 sqft and the units of price are 1000s of dollars.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

Creating `x_train` and `y_train` variables in one dimensional NumPy array.

```python
# x_train is the input variable 
# y_train is the target 
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

Number of training samples `m`

```python
print(f"X_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training samples is: {m}")
```

#### Training 
You will use $(x^{(i)}, y^{(i)})$ to denote the $i^{th}$ training example. 


```python
i = 1

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

#### Plotting the data

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("House Prices")
plt.ylabel("Price (in 100s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show()
```

#### Model Function
The model function for linear regression (which is a function that maps from `x` to `y`) is represented as 

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

This formula above is how you can represent straight lines - different values of $w$ and $b$ give you different straight lines on the plot. <br/> <br/> <br/> <br/> <br/> 

starting with $w = 100$ and $b = 100$. 

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

Computing the value of $f_{w,b}(x^{(i)})$ for your two data points. 

for $x^{(0)}$, $$f_wb = w * x[0] + b$$

for $x^{(1)}$, $$f_wb = w * x[1] + b$$



```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
        x(ndarray (m,)): Data, m examples
        w, b (scaler)  : Model Parameters
    Return:
    f_wb (ndarray (m,)): model Predication
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

plotting the plot for the given values

```python
tmp_f_wb = compute_model_output(x_train, w, b)

# plotting the model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

# plotting the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()


```
